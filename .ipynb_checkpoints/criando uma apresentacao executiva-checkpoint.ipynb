## Importando Bibliotecas e visualizando o conjunto de dados

### Venda por ano

```python
arquivo['Year'] = arquivo['Order Date'].dt.year
```

```python
arquivo.groupby("Year")["Sales"].sum().plot.bar();
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(arquivo.groupby("Year")["Sales"].sum().index, arquivo.groupby("Year")["Sales"].sum().values)
ax.set_title("Vendas por ano",fontsize=14, fontweight="bold")
ax.xaxis.set_ticks([2015,2016,2017,2018])
plt.show()
```

```python
arquivo["Month"] = arquivo["Order Date"].dt.month
```

```python
arquivo.head(2)
```

```python
soma_mes = arquivo.groupby(["Month","Year"])[("Sales")].sum()
```

```python
soma_mes = soma_mes.reset_index()
```

```python
soma_mes
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2015].Month.values, soma_mes[soma_mes.Year == 2015].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2016].Month.values, soma_mes[soma_mes.Year == 2016].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2017].Month.values, soma_mes[soma_mes.Year == 2018].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2018].Month.values, soma_mes[soma_mes.Year == 2018].Sales.values)
plt.show()
```

### Simplificando visualização dos gráficos

```python
fig,ax = plt.subplots(
            figsize = (12,8) # Tamanho da figura
        )

wid = 0.2

ax.bar(soma_mes[soma_mes.Year == 2015].Month.values-wid-wid/2, # <- DESLOCANDO A BARRA PARA A ESQUERDA
       soma_mes[soma_mes.Year == 2015].Sales.values,
       color="blue", # Cor das barras
       label=2015, # Label dos dados
       width=wid # Largura das barras
      )
ax.bar(soma_mes[soma_mes.Year == 2016].Month.values-wid/2, # <- DESLOCANDO A BARRA PARA A DIREITA
       soma_mes[soma_mes.Year == 2016].Sales.values,
       color="green", 
       label=2016,
       width=wid 
      )
ax.bar(soma_mes[soma_mes.Year == 2017].Month.values+wid/2, # <- DESLOCANDO A BARRA PARA A ESQUERDA
       soma_mes[soma_mes.Year == 2017].Sales.values,
       color="brown", # Cor das barras
       label=2017, # Label dos dados
       width=wid # Largura das barras
      )
ax.bar(soma_mes[soma_mes.Year == 2018].Month.values+wid+wid/2, # <- DESLOCANDO A BARRA PARA A DIREITA
       soma_mes[soma_mes.Year == 2018].Sales.values,
       color="red", 
       label=2018,
       width=wid 
      )

ax.legend()

plt.show()
```

### Qual foi a categoria mais vendida?


```python
arquivo.head(2)
```

```python
arquivo.groupby("Category")["Sales"].sum()
```

```python
categoria = arquivo.groupby(["Category", "Year"])["Sales"].sum()
```

```python
categoria.head()
```

```python
categoria = categoria.reset_index()
```

```python
categoria.head()
```

```python
fig,ax = plt.subplots(
            figsize = (16,4) 
        )

wid = 0.2 

ax.bar(categoria[categoria.Category == 'Furniture'].Year.values-wid, 
       categoria[categoria.Category == 'Furniture'].Sales.values,
       color="#00c698", 
       label='Furniture', 
       width=wid 
      )
ax.bar(categoria[categoria.Category == 'Office Supplies'].Year.values, 
       categoria[categoria.Category == 'Office Supplies'].Sales.values,
       color="#00a9b5", 
       label='Office Supplies', 
       width=wid 
      )
ax.bar(categoria[categoria.Category == 'Technology'].Year.values+wid, 
       categoria[categoria.Category == 'Technology'].Sales.values,
       color="#008ac5", 
       label='Technology', 
       width=wid 
      )

ax.legend()


for i in np.arange(0,4):
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Furniture'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Furniture'].Year.values[i]-wid,categoria[categoria.Category == 'Furniture'].Sales.values[i]), 
                ha="center",  
                va="top", 
                xytext=(0,12), 
                textcoords="offset points", 
                fontsize=10,
                fontweight='bold', 
                color="#00c698" 
            )
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Office Supplies'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Office Supplies'].Year.values[i],categoria[categoria.Category == 'Office Supplies'].Sales.values[i]), 
                ha="center", 
                va="top", 
                xytext=(0,12), 
                textcoords="offset points", 
                fontsize=10, 
                fontweight='bold', 
                color="#00a9b5" 
            )
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Technology'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Technology'].Year.values[i]+wid,categoria[categoria.Category == 'Technology'].Sales.values[i]), 
                ha="center",  
                va="top", 
                xytext=(0,12),
                textcoords="offset points", 
                fontsize=10, 
                fontweight='bold',
                color="#008ac5"
            )
ax.set_yticks(np.array([0,320000])) 
ax.yaxis.set_visible(False) 
    
ax.spines['top'].set_visible(False) 
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.xaxis.set_ticks([2015,2016,2017,2018]) 
ax.tick_params(axis='x',labelsize=14) 

plt.show()
```

```python
itens = arquivo.groupby("Product Name")["Sales"].sum()
itens = itens.reset_index()
itens = itens.sort_values("Sales",ascending=False)
itens.head(10)
```

```python
n = 10
top_n_itens = itens.head(n)
top_n_itens
```

```python
fig,ax = plt.subplots(
            figsize = (16,12)
        )

ax.barh(top_n_itens['Product Name'].values,
        top_n_itens['Sales'].values,  
        align='center')

ax.invert_yaxis() 

for i in np.arange(0,n):
    ax.annotate('{:,.0f}k'.format(top_n_itens['Sales'].values[i]/1000), 
                (top_n_itens['Sales'].values[i],i), 
                ha="center", va="top", xytext=(-15,3), textcoords="offset points", 
                fontsize=12, fontweight='bold', color="white")
    
ax.xaxis.set_visible(False) 

ax.spines['top'].set_visible(False) 
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.show()
```

### Juntando base no pandas

```python
# Criando 2 dataframes
dic1 = {
    "names": ['Name1','Name2','Name3'],
    "values": [1,2,3]
}

base_dic1 = pd.DataFrame(dic1)

dic2 = {
    "names": ['Name1','Name2','Name4'],
    "values": [9,8,7]
}

base_dic2 = pd.DataFrame(dic2)
```

```python
arquivo_merge = pd.merge(
    base_dic1, 
    base_dic2, 
    how='right', 
    on="names"
) 

display(arquivo_merge)
```

```python
df_datas = pd.DataFrame([2018,2017,2016,2015])
df_datas.columns = ['Year']
df_datas
```

```python
df_datas['uniao'] = 'unir'
itens['uniao'] = 'unir'
```

```python
n = 10
top_n_itens = itens.head(n)
```

```python
produto_ano = pd.merge(
    top_n_itens, # <- primeira base
    df_datas, # <- segunda base
    how='outer', # <- tipo de junção que vamos fazer
    on="uniao" # <- coluna que vamos usar para fazer essa junção das bases
) 

produto_ano.head()
```

```python
itens_ano_grafico = pd.merge(
    produto_ano,
    itens_ano,
    how='left',
    on=['Product Name','Year']
)
        
itens_ano_grafico.head()
```

```python
itens_ano_grafico[itens_ano_grafico.Year == 2018].Sales_y.values
```

```python
fig,ax = plt.subplots(
            figsize = (16,12)
        )

hgt = 0.2

qtd_itens = len(top_n_itens)

ax.barh(np.arange(0,qtd_itens)-hgt-hgt/2, 
        itens_ano_grafico[itens_ano_grafico.Year == 2018].Sales_y.values, 
        align='center',
        height=hgt, 
        label=2018,
        color="#0069c0"
       )
ax.barh(np.arange(0,qtd_itens)-hgt/2, 
        itens_ano_grafico[itens_ano_grafico.Year == 2017].Sales_y.values, 
        align='center',height=hgt,label=2017,color="#008ac5")
ax.barh(np.arange(0,qtd_itens)+hgt/2, 
        itens_ano_grafico[itens_ano_grafico.Year == 2016].Sales_y.values,
        align='center',height=hgt,label=2016,color="#00a9b5")
ax.barh(np.arange(0,qtd_itens)+hgt+hgt/2, 
        itens_ano_grafico[itens_ano_grafico.Year == 2015].Sales_y.values, 
        align='center',height=hgt,label=2015,color="#00c698")

ax.invert_yaxis() 
ax.legend()

for i in np.arange(0,qtd_itens):
    ax.annotate('{:,.1f}k'.format(itens_ano_grafico[itens_ano_grafico.Year == 2018].Sales_y.values[i]/1000), 
                (itens_ano_grafico[itens_ano_grafico.Year == 2018].Sales_y.values[i],i-hgt-hgt/2), 
                ha="center", va="top", xytext=(20,6), textcoords="offset points", 
                fontsize=12, fontweight='bold', color="#0069c0")
    
ax.yaxis.set_ticks(np.arange(0,qtd_itens)) 
ax.set_yticklabels(top_n_itens["Product Name"])
ax.tick_params(axis='y',labelsize=10)

ax.xaxis.set_visible(False) 

ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.show()
```
