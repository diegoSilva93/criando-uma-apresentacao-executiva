### Venda por ano

```python
arquivo['Year'] = arquivo['Order Date'].dt.year
```

```python
arquivo.groupby("Year")["Sales"].sum().plot.bar();
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(arquivo.groupby("Year")["Sales"].sum().index, arquivo.groupby("Year")["Sales"].sum().values)
ax.set_title("Vendas por ano",fontsize=14, fontweight="bold")
ax.xaxis.set_ticks([2015,2016,2017,2018])
plt.show()
```

```python
arquivo["Month"] = arquivo["Order Date"].dt.month
```

```python
arquivo.head(2)
```

```python
soma_mes = arquivo.groupby(["Month","Year"])[("Sales")].sum()
```
