### Venda por ano

```python
arquivo['Year'] = arquivo['Order Date'].dt.year
```

```python
arquivo.groupby("Year")["Sales"].sum().plot.bar();
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(arquivo.groupby("Year")["Sales"].sum().index, arquivo.groupby("Year")["Sales"].sum().values)
ax.set_title("Vendas por ano",fontsize=14, fontweight="bold")
ax.xaxis.set_ticks([2015,2016,2017,2018])
plt.show()
```

```python
arquivo["Month"] = arquivo["Order Date"].dt.month
```

```python
arquivo.head(2)
```

```python
soma_mes = arquivo.groupby(["Month","Year"])[("Sales")].sum()
```

```python
soma_mes = soma_mes.reset_index()
```

```python
soma_mes
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2015].Month.values, soma_mes[soma_mes.Year == 2015].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2016].Month.values, soma_mes[soma_mes.Year == 2016].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2017].Month.values, soma_mes[soma_mes.Year == 2018].Sales.values)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.bar(soma_mes[soma_mes.Year == 2018].Month.values, soma_mes[soma_mes.Year == 2018].Sales.values)
plt.show()
```

### Simplificando visualização dos gráficos

```python
fig,ax = plt.subplots(
            figsize = (12,8) # Tamanho da figura
        )

wid = 0.2

ax.bar(soma_mes[soma_mes.Year == 2015].Month.values-wid-wid/2, # <- DESLOCANDO A BARRA PARA A ESQUERDA
       soma_mes[soma_mes.Year == 2015].Sales.values,
       color="blue", # Cor das barras
       label=2015, # Label dos dados
       width=wid # Largura das barras
      )
ax.bar(soma_mes[soma_mes.Year == 2016].Month.values-wid/2, # <- DESLOCANDO A BARRA PARA A DIREITA
       soma_mes[soma_mes.Year == 2016].Sales.values,
       color="green", 
       label=2016,
       width=wid 
      )
ax.bar(soma_mes[soma_mes.Year == 2017].Month.values+wid/2, # <- DESLOCANDO A BARRA PARA A ESQUERDA
       soma_mes[soma_mes.Year == 2017].Sales.values,
       color="brown", # Cor das barras
       label=2017, # Label dos dados
       width=wid # Largura das barras
      )
ax.bar(soma_mes[soma_mes.Year == 2018].Month.values+wid+wid/2, # <- DESLOCANDO A BARRA PARA A DIREITA
       soma_mes[soma_mes.Year == 2018].Sales.values,
       color="red", 
       label=2018,
       width=wid 
      )

ax.legend()

plt.show()
```

### Qual foi a categoria mais vendida?


```python
arquivo.head(2)
```

```python
arquivo.groupby("Category")["Sales"].sum()
```

```python
categoria = arquivo.groupby(["Category", "Year"])["Sales"].sum()
```

```python
categoria.head()
```

```python
categoria = categoria.reset_index()
```

```python
categoria.head()
```

```python
# Vamos utilizar o mesmo código do gráfico anterior para exibir por categorias
fig,ax = plt.subplots(
            figsize = (16,4) # Tamanho da figura
        )

wid = 0.2 # Tamanho das barras

ax.bar(categoria[categoria.Category == 'Furniture'].Year.values-wid, 
       categoria[categoria.Category == 'Furniture'].Sales.values,
       color="#00c698", # Cor das barras
       label='Furniture', # Label dos dados
       width=wid # Largura das barras
      )
ax.bar(categoria[categoria.Category == 'Office Supplies'].Year.values, 
       categoria[categoria.Category == 'Office Supplies'].Sales.values,
       color="#00a9b5", # Cor 
       label='Office Supplies', 
       width=wid 
      )
ax.bar(categoria[categoria.Category == 'Technology'].Year.values+wid, 
       categoria[categoria.Category == 'Technology'].Sales.values,
       color="#008ac5", 
       label='Technology', 
       width=wid 
      )

ax.legend()


for i in np.arange(0,4):
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Furniture'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Furniture'].Year.values[i]-wid,categoria[categoria.Category == 'Furniture'].Sales.values[i]), 
                ha="center",  
                va="top", 
                xytext=(0,12), 
                textcoords="offset points", 
                fontsize=10,
                fontweight='bold', 
                color="#00c698" 
            )
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Office Supplies'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Office Supplies'].Year.values[i],categoria[categoria.Category == 'Office Supplies'].Sales.values[i]), 
                ha="center", 
                va="top", 
                xytext=(0,12), 
                textcoords="offset points", 
                fontsize=10, 
                fontweight='bold', 
                color="#00a9b5" 
            )
    ax.annotate('{:,.0f}k'.format(categoria[categoria.Category == 'Technology'].Sales.values[i]/1000), 
                (categoria[categoria.Category == 'Technology'].Year.values[i]+wid,categoria[categoria.Category == 'Technology'].Sales.values[i]), 
                ha="center",  
                va="top", 
                xytext=(0,12),
                textcoords="offset points", 
                fontsize=10, 
                fontweight='bold',
                color="#008ac5"
            )
ax.set_yticks(np.array([0,320000])) 
ax.yaxis.set_visible(False) 
    
ax.spines['top'].set_visible(False) 
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.xaxis.set_ticks([2015,2016,2017,2018]) 
ax.tick_params(axis='x',labelsize=14) 

plt.show()
```
